```python
import torch
import cv2
import numpy as np
```

```python
import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath
```

```python
model = torch.hub.load("ultralytics/yolov5" , "custom", path = "best.pt", force_reload = True)
```

```python
img = cv2.imread(r"C:\Users\raeha\Downloads\img6.jpg")



result = model(img)

img = result.render()[0]
img = cv2.resize(img, (400, 490)) 
cv2.imshow("body",img)
cv2.waitKey(0)
```

```python
cap = cv2.VideoCapture(r"C:\Users\raeha\Downloads\vid2.mp4") 
while True:
    ret, frame = cap.read()
    if not ret:
        break

    result = model(frame)

    frame = result.render()[0]

    frame = cv2.resize(frame, (340, 480))  

    # اعرض الفريم
    cv2.imshow("body", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
